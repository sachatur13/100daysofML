```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image = cv2.imread(r'scan.jpg')
```

```python
plt.imshow(input_image)
```

```python
## getting coordinates
p_A = np.float32([[332,15],[700,215],[88,600],[500,700]])
p_B = np.float32([[0,0],[420,0],[0,600],[420,600]])

transformed = cv2.getPerspectiveTransform(p_A,p_B)
warped = cv2.warpPerspective(input_image,transformed,(420,594))
plt.imshow(warped)
```
