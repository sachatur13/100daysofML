```python
## Scanning, Finding edges etc
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image = cv2.imread(r'C:\Users\saket\Desktop\courses\TF\OpenCV\Master OpenCV\images\Origin_of_Species.jpg',0)
```

```python
plt.imshow(input_image)
```

```python
ret,thresh1 = cv2.threshold(input_image,125,255,cv2.THRESH_BINARY)
plt.imshow(thresh1)
```

```python
ret,thresh1 = cv2.threshold(input_image,125,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh1)
```

```python
ret,thresh1 = cv2.threshold(input_image,125,255,cv2.THRESH_TRUNC)
plt.imshow(thresh1)
```

```python
ret,thresh1 = cv2.threshold(input_image,125,255,cv2.THRESH_TOZERO)
plt.imshow(thresh1)
```

```python
ret,thresh1 = cv2.threshold(input_image,125,255,cv2.THRESH_TOZERO_INV)
plt.imshow(thresh1)
```

```python
thresh1 = cv2.adaptiveThreshold(input_image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,3,5)

plt.imshow(thresh1)
```

```python
_,thresh1 = cv2.threshold(input_image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)


plt.imshow(thresh1)
```
