```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image = cv2.imread(r'Untitled.jpg',0)
```

```python
plt.imshow(input_image)
```

```python
## binarize image using threshold
ret,thresh = cv2.threshold(input_image.copy(),127,255,cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(thresh)
```

```python
## find contours
cont,hier = cv2.findContours(thresh,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
```

```python
## loop and find all
plt.figure(figsize=(10,12))
for c in cont:
    x,y,w,h = cv2.boundingRect(c)
    cv2.rectangle(input_image,(x,y),(x+w,y+h),(0,255,0),2)
    plt.imshow(input_image)
```

```python
## Convex Hull -- Adding the bounding box in a polygon
input_image_2 = cv2.imread(r'Capture.JPG')
```

```python
plt.imshow(input_image_2)
```

```python
bgr = cv2.cvtColor(input_image_2,cv2.COLOR_BGR2GRAY)
plt.imshow(bgr)
```

```python
## Threshold
ret,thresh = cv2.threshold(bgr,100,88,0)
```

```python
plt.imshow(thresh)
```

```python
contour,h = cv2.findContours(thresh.copy(),cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
```

```python
## loop and find all
plt.figure(figsize=(10,12))
for c in contour:
    x,y,w,h = cv2.boundingRect(c)
    cv2.rectangle(input_image_2,(x,y),(x+w,y+h),(0,255,0),2)
    plt.imshow(input_image_2)
```

```python
## Sort and remove largest
n = len(contour)-1
contours = sorted(contour,key = cv2.contourArea,reverse = False)[:n]

## Iterate and draw convex hull
plt.figure(figsize=(10,12))
for c in contours:
    hull = cv2.convexHull(c)
    cv2.drawContours(input_image_2,[hull],0,(0,255,0),2)
    plt.imshow(input_image_2)
```
