```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image = cv2.imread(r'C:\Users\saket\Desktop\courses\TF\OpenCV\Master OpenCV\images\truck.jpg                         ')
```

```python
plt.imshow(input_image)
```

```python
gray = cv2.cvtColor(input_image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
ORB = cv2.ORB_create(10)
```

```python
keypoints = ORB.detect(gray,None)
```

```python
key,descriptor = ORB.compute(gray,keypoints)
print(len(key))
```

```python
input_image = cv2.drawKeypoints(input_image,keypoints,input_image,flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(input_image)
```
