```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image = cv2.imread(r'abraham.jpg')
```

```python
plt.imshow(input_image)
```

```python
New_Matrix = np.ones(input_image.shape,dtype = 'uint8')*100

## Adding the matrix to the image

new_image = cv2.add(input_image,New_Matrix)
plt.imshow(new_image)
```

```python
## Adding the matrix to the image
new_image = cv2.subtract(input_image,New_Matrix)
plt.imshow(new_image)
```
