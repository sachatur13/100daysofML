```python
import numpy as np
import pandas as pd
from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
```

```python
(train_images,train_lables),(test_images,test_labels) = fashion_mnist.load_data()
```

```python
train_images.shape
```

```python
from tensorflow.keras.utils import to_categorical

train_lables_categorical = to_categorical(train_lables)
```

```python
plt.imshow(train_images[1])
plt.colorbar()
```

```python
## Preprocessing images 
train_images = train_images/255
test_images = test_images/255
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,Flatten
```

```python
model = Sequential()

model.add(Flatten(input_shape = (28,28)))
model.add(Dense(264,activation='relu'))
model.add(Dense(128,activation = 'relu'))
model.add(Dense(10))

model.compile(optimizer = 'adam',loss = 'categorical_crossentropy',metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model_fit = model.fit(train_images,train_lables_categorical,validation_split = 0.22, epochs = 15,batch_size = 32)
```
