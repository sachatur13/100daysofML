```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image = cv2.imread(r'chess.jpg')
```

```python
plt.imshow(input_image)
```

```python
gray = cv2.cvtColor(input_image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
gray = np.float32(gray)
```

```python
## Corner detection using Harris Corner algorithm
harris = cv2.cornerHarris(gray,3,3,0.05) ## Returns array of corner location
```

```python
## Enlarging points on the corners
kernel = np.ones((7,7),np.uint8)
har = cv2.dilate(harris,kernel,iterations = 5)
```

```python
input_image[harris>0.025 * harris.max()] = [255,127,127]
```

```python
plt.figure(figsize  = (12,10))
plt.imshow(input_image)
```

```python
cv2.imshow('harris',input_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## Good Features to Track method

corners = cv2.goodFeaturesToTrack(gray,50,0.01,15)
for corner in corners:
    x,y = corner[0]
    x = int(x)
    y = int(y)
    
    cv2.rectangle(input_image,(x-10,y-10),(x+10,y+10),(0,255,0),2)

plt.figure(figsize = (10,12))
plt.imshow(input_image)
```
