```python
## Translations
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread(r'input.jpg')
height,width = image.shape[:2]

## Creating the translation matrix
q_height,q_width = height/4,width/4

T = np.float32([[1,0,q_height],[0,1,q_width]])

translation = cv2.warpAffine(image,T,(width,height))
#cv2.imshow('translated', translation)
plt.imshow(translation)

#cv2.waitKey()
#cv2.destroyAllWindows()
```

```python
## Rotations in the image

rotation_matrix = cv2.getRotationMatrix2D((width/3,height/2),90,1)
rotated = cv2.warpAffine(image,rotation_matrix,(width,height))
plt.imshow(rotated)
```

```python
## Scaling Resizing and interpolation
scaled = cv2.resize(image,None,fx =0.5,fy=0.5)
plt.imshow(image)
```

```python
##increasing
plt.imshow(cv2.resize(image,None,fx=5,fy=3,interpolation = cv2.INTER_AREA))
```

```python
## Image pyramids
## cv2.pyrDown,, cv2.pyrUp

smaller = cv2.pyrDown(image)
larger = cv2.pyrUp(image)
cv2.imshow('xx',smaller)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## Cropping and image
## Getting the box to crop
start_row,start_col = int(height*0.2),int(width*0.2)
end_row,end_col = int(height*0.5),int(width*0.5)

## using indexing to crop the image
cropped = image[start_row:end_row,start_col:end_col]
plt.imshow(cropped)
```
