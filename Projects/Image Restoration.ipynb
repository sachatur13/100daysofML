```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image = cv2.imread(r'C:\Users\saket\Desktop\courses\TF\OpenCV\Master OpenCV\images\joker.jpg')
plt.imshow(input_image)
```

```python
marked_image = cv2.imread(r'C:\Users\saket\Desktop\courses\TF\OpenCV\Master OpenCV\images\joker2.jpg',0)
plt.imshow(marked_image)
```

```python
## Thresholding
ret,thresh = cv2.threshold(marked_image,254,255,cv2.THRESH_BINARY)
plt.imshow(thresh)
```

```python
## Dilation
kernel = np.ones((7,7),np.uint8)
mask = cv2.dilate(thresh,kernel,iterations = 1)
plt.imshow(mask)
```

```python
## Restor
rest = cv2.inpaint(input_image,mask,3,cv2.INPAINT_TELEA)
plt.imshow(rest)
```
