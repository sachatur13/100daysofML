```python
## Image path setting
image_path = os.listdir(train_data)[0]
```

```python
## Show image file
face = train_data+image_path

def read_image_from_path(image_path):
    print(image_path)
    if os.path.isfile(image_path):
        faceimage = cv2.imread(image_path,flags = 0)
        plt.imshow(cv2.imread(image_path,flags =0),cmap = 'gray')
    else:
        print("File does not exist")
    
    return faceimage
```
