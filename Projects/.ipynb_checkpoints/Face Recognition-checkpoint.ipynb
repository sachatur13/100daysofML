```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Loading HAAR classifier
classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def extractface(input_image):
    ## Return cropped face
    gray = cv2.cvtColor(input_image,cv2.COLOR_BGR2GRAY)
    face = classifier.detectMultiScale(gray,1.3,5)

    if face is ():
        return None
    
    for (x,y,w,h) in face:
        cropped = input_image[y:y+h,x:x+w]
    return cropped
```

```python
capture_face = cv2.VideoCapture(0)
i = 0 ## counter for face value

## Collect sample
while True:
    ret,frame = capture_face.read()
    if extractface(frame) is not None:
        i+=1
        face = cv2.resize(extractface(frame),(200,200))
        face = cv2.cvtColor(face,cv2.COLOR_BGR2GRAY)

        ## Save training images
        path = r'C:\Users\saket\Desktop\courses\TF\OpenCV\Master OpenCV\faces\Saket\'' +str(i)+ '.jpg'
        cv2.imwrite(path,face)

        ## counts
        cv2.putText(face,str(i),(50,50),cv2.FONT_HERSHEY_COMPLEX_SMALL,1,(0,255,0),2)
        plt.imshow(face)
    
    else:
        print('No Face')
        pass

    if cv2.waitKey(1)==13 or i == 100:
        break
```

```python
capture_face.release()
```
